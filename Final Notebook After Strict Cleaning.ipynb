According to the data information, 'ConnectedToGrid' and 'CutMeals' hold integer values, we therefore have no reason to let then be floats. Also it makes logical sense not to have a half bedrooms. Hence we'll take these three columns to int64s

```python
df['ConnectedToGrid'] = df['ConnectedToGrid'].astype(int)
df['Bedrooms'] = df['Bedrooms'].astype(int)
df['CutMeals'] = df['CutMeals'].astype(int)
```

```python
df.dtypes
```

```python
variable = (df == 0).astype(int).sum(axis=1)
```

```python
NewTable = pd.DataFrame(variable)
```
